```python
import cv2
import sys
import pytesseract
from matplotlib import pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
data_path = os.path.join("datasets", "")
```

## English Text

```python

# Read image path from command line
img_path = data_path + '/scans/img_eng.jpg'

# Define config parameters.
# '-l' English language
# '--oem' for specifying Engine (LSTM or not...)
config = ('-l eng --oem 1 --psm 3')

# Read image from disk
im = cv2.imread(img_path, cv2.IMREAD_COLOR)

#plot image
plt.imshow(im)

# Run tesseract OCR on image
text = pytesseract.image_to_string(im, config=config)

# Print recognized text
print(text)
```

## Arabic Text

### Convert PDF to Jpeg

```python
from pdf2image import convert_from_path

pdf_path = data_path +'DENONCIATIONS/ABCDetGHI.pdf'
img_path = data_path +'scans/img.jpg'

pages = convert_from_path(pdf_path, 500)

for page in pages:
    page.save(img_path, 'JPEG')
```

```python
config = ('-l ara --oem 1 --psm 3')

im = cv2.imread(img_path, cv2.IMREAD_COLOR)

#plt.imshow(im)

text = pytesseract.image_to_string(im, config=config)

print(text)
```
