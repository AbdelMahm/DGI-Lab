```python
import cv2
import sys
import pytesseract
from matplotlib import pyplot as plt
import numpy as np
```

## English Text

```python
# Read image path from command line
img_path = 'images/img_eng.jpg'

# Define config parameters.
# '-l' English language
# '--oem' for specifying Engine (LSTM or not...)
config = ('-l ara --oem 1 --psm 3')

# Read image from disk
im = cv2.imread(img_path, cv2.IMREAD_COLOR)

#plot image
plt.imshow(im)

# Run tesseract OCR on image
text = pytesseract.image_to_string(im, config=config)

# Print recognized text
print(text)
```

## Arabic Text

```python
# Requirements: tested on Mac OS
# must set the variable 'export TESSDATA_PREFIX=/usr/local/Cellar/tesseract/HEAD-e9db7aa/share/tessdata/'
# must download 'ara.traineddata' form 'https://github.com/tesseract-ocr/tessdata/raw/master/ara.traineddata'
# and put it in TESSDATA_PREFIX
```

```python
# Read image path from command line
img_path = 'images/img_ara.png'

config = ('-l ara --oem 1 --psm 3')

im = cv2.imread(img_path, cv2.IMREAD_COLOR)

plt.imshow(im)

text = pytesseract.image_to_string(im, config=config)

print(text)
```

## French Texte

```python
# must download 'fra.traineddata' form 'https://github.com/tesseract-ocr/tessdata/raw/master/fra.traineddata'
# and put it in TESSDATA_PREFIX
```

```python
# Read image path from command line
img_path = 'images/img_fra.png'

config = ('-l fra --oem 1 --psm 3')

im = cv2.imread(img_path, cv2.IMREAD_COLOR)


plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.imshow(im)

text = pytesseract.image_to_string(im, config=config)

print(text)
```
